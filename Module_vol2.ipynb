## Розв'язок

Швидше за все, для розв'язання задачі нам потібен доступ до камери (або, наприклад, знімки із супутника, можливо) щоб розуміти стан і загруженість ділянки дороги, яку регулює світлофор. На мою думку, це очевидний і в той же час ефективний підхід.

Оскільки будемо працювати із зображеннями, потрібні методи розпізнавання образів (машини в нашому випадку). Із урахуванням того, що із сучасними иехнологіями ця задача не потребує особливих технічних ресурсів, навіть за використання Глибинних Нейронних мереж, можемо використовувати будь-яку із уже існуючих, навіть заздалегідь натренованих мереж (YOLO), для розпізнавання образів машин.

Далі, коли ми знаємо де і скільки знаходиться машин, необзідно прийняти рішеннь, яке світло на світлофорі має горіти. Я думаю, що для цього можна звернутися до методів **Reinforcement Learning**, оскільки він підходить під  поставлену задачу: 
> **Reinforcement learning (RL)** is an area of machine learning concerned with how intelligent agents ought to take actions in an environment in order to maximize the notion of cumulative reward.

>Reinforcement learning differs from supervised learning in  ***not needing labelled input/output pairs*** be presented, and in not needing sub-optimal actions to be explicitly corrected. Instead the focus is on finding a balance between exploration (of uncharted territory) and exploitation (of current knowledge)

Тобто в такому випадку нам навіть не потрібно початковий датасет, для тренування, лише певний набір критеріїв, що визначав би, наскільки ефективним було прийняте алгоритмом рішення, і надалі ці результати можуть використовуввтися цією ж мережею. (На початку роботи алгоритму можна щоб світловор перемикався за таймером, як звичайно, а потім вже коригувати його рішення)

*Я на жаль раніше не працювала тісно із алгоритмами цього напряму, але думаю, що це правильний напрям у якому треба дивитися для розв'язання проблеми. Щоб детальніше відповісти на запитання, яку архітектуру мережі слід обрати, треба детальші ознайомитися із існуючими методами, а для цього потрібно більше часу.*
